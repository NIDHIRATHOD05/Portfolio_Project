```python
import pandas as pd 
import seaborn as sns #
import numpy as np


import matplotlib # for drafting graphs 
import matplotlib.pyplot as plt
```

```python
movie_data = pd.read_csv('/Users/nidhirathod/Downloads/movies.csv')
```

```python
movie_data.head()
```

```python
movie_data.head()
```

```python
# Let's look for missing data first 

for column in movie_data:
    null_values = np.mean(movie_data[column].isnull())
    print('{} - {}%' .format(column , null_values))
```

```python
# datatype for columns

movie_data.dtypes
```

```python
#filling in the null


movie_data["budget"].fillna(value="00" , inplace= True)

movie_data["gross"].fillna(value="00" , inplace= True)

movie_data["votes"].fillna(value="00" , inplace= True)

movie_data["runtime"].fillna(value="00" , inplace= True)



movie_data.head()
```

```python
# changing the datatype

movie_data['budget'] = movie_data['budget'].astype('int64')

movie_data['gross'] = movie_data['gross'].astype('int64')

movie_data['votes'] = movie_data['votes'].astype('int64')

movie_data['runtime'] = movie_data['runtime'].astype('int64')

movie_data.head()
```

```python
# split the values of released


movie_data["released"].str.split(", " ,expand=True) 

movie_data.head()


```

```python
#Spliting the values into different columns 

movie_data[["Date" , "Year_State"]] = movie_data["released"].str.split(", " , expand=True)


movie_data.head()
```

```python
movie_data["Year_State"].astype(str).str[:4]
movie_data.head()
```

```python
movie_data["new_year"] = movie_data["Year_State"].astype(str).str[:4]

movie_data.head()
```

```python
movie_data = movie_data.sort_values(by=["gross"] , inplace=False , ascending=False)

pd.set_option('display.max_rows', 50)
```

```python
movie_data
```

```python
# DROP THE DUPLICATE VALUES

movie_data["company"].drop_duplicates().sort_values(ascending=False)
```

```python
movie_data.drop_duplicates()
```

```python
# Correlations --> Budget and Gross
# Scatter plot with budget v/s gross 



plt.scatter(x = movie_data["budget"] , y = movie_data["gross"])

plt.title('Budget v/s Gross Earning')

plt.xlabel('Gross Earning')

plt.ylabel('Budget for Flim')


plt.show()
```

```python
movie_data.head()
```

```python
# Regerration Plot , Plot budget v/s gross using seaborn

sns.regplot(x='budget' , y='gross' , data=movie_data, scatter_kws={"color" : "red"}, line_kws={"color" : "blue"})
```

```python
# Let's star looking at Correlations --> Pearson, kandall , spearman 

movie_data.corr(method='pearson')
```

```python
correlation_matrix = movie_data.corr(method='pearson')

sns.heatmap(correlation_matrix , annot=True) # annot = annotation -->a way to associate arbitrary expressions to function arguments and return values.

plt.title('Correlation Matrix for Numeric Feature')

plt.xlabel('Movie Feature')

plt.ylabel('Movie Feature')


plt.show()
```

```python
numerical_data = movie_data

for column_name in numerical_data.columns:
    if(numerical_data[column_name].dtype == 'object'):
        numerical_data[column_name] = numerical_data[column_name].astype('category')
        numerical_data[column_name] = numerical_data[column_name].cat.codes

numerical_data.head()

```

```python
correlation_matrix = numerical_data.corr(method='pearson')

sns.heatmap(correlation_matrix , annot=True) # annot = annotation -->a way to associate arbitrary expressions to function arguments and return values.

plt.title('Correlation Matrix for Numeric Feature')

plt.xlabel('Movie Feature')

plt.ylabel('Movie Feature')


plt.show()
```

```python
numerical_data.corr()
```

```python
matrix = numerical_data.corr()

corr_pair = matrix.unstack()

corr_pair
```

```python
sorted_pairs = corr_pair.sort_values()

sorted_pairs
```

```python
highest_corr = sorted_pairs[(sorted_pairs) > 0.5]

highest_corr
```
