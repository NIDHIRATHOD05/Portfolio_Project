Applying a Function to every row:



```python
netflix["genre_count"] = netflix["listed_in"].map(lambda x : len(x.split(","))) # lambda x = function , len = give the length of the list and perfrom the operation provided
netflix                                                                         # x.split = split the value of each elements of the list (listed_by) by comma .
```

Mostly we use map() function:



```python
netflix["nonsensical_value"] = netflix.apply(lambda x : len(x["cast"].split(",")) + len(x["listed_in"].split(",")), axis=1)
netflix
```

```python
netflix.drop(columns="nonsensical_value")
```
